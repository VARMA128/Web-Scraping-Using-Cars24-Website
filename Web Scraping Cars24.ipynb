```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
website="https://www.cars24.com/buy-used-car/?sort=bestmatch&serveWarrantyCount=true&listingSource=Homepage_Filters&storeCityId=3686"
```

```python
response=requests.get(website)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
car_year= soup.find_all(class_="sc-braxZu kjFjan")
```

```python

for Car_year in car_year:
    print(Car_year.get_text())
```

```python
fuel_type= soup.find_all('p',class_="sc-braxZu kvfdZL")
```

```python

for Fuel_type in fuel_type:
    print(Fuel_type.get_text())
```

```python

car_names = []
years = []

for Car_year in car_year:
    full_text = Car_year.get_text().strip()
    parts = full_text.split(' ', 1)  # Split into year and car name

    if len(parts) == 2:
        year = parts[0]
        car = parts[1]
        car_names.append(car)
        years.append(year)
```

```python
for car in car_names:
    print(car)
```

```python
for year in years:
    print(year)
```

```python
texts = [item.get_text(strip=True) for item in fuel_type]

km_list = []
fuel_list = []
trans_list = []

for i in range(0, len(texts), 5):
    block = texts[i:i+5]
    if len(block) >= 3:
        km_list.append(block[0])
        fuel_list.append(block[1])
        trans_list.append(block[2])
```

```python
for km in km_list:
    print(km)
```

```python
for fuel in fuel_list:
    print(fuel)
```

```python
for trans in trans_list:
    print(trans)
```

```python
price_list = []

for Price in price:
    price_list.append(Price.get_text(strip=True))

for p in price_list:
    print(p)
```

```python

car_names = []
years = []
km_list = []
fuel_list = []
trans_list = []
price_list = []

car_year_elements = soup.find_all("span", class_="sc-braxZu kjFjan")
for item in car_year_elements:
    text = item.get_text(strip=True)
    parts = text.split(' ', 1)
    if len(parts) == 2:
        years.append(parts[0])
        car_names.append(parts[1])

specs = soup.find_all("p", class_="sc-braxZu kvfdZL")
spec_texts = [s.get_text(strip=True) for s in specs]

for i in range(0, len(spec_texts), 5):
    block = spec_texts[i:i+5]
    if len(block) >= 3:
        km_list.append(block[0])
        fuel_list.append(block[1])
        trans_list.append(block[2])

prices = soup.find_all("p", class_="sc-braxZu cyPhJl")
for p in prices:
    price_list.append(p.get_text(strip=True))

final_len = min(len(car_names), len(years), len(km_list), len(fuel_list), len(trans_list), len(price_list))

data = {
    "car_name": car_names[:final_len],
    "year": years[:final_len],
    "kilometers": km_list[:final_len],
    "fuel_type": fuel_list[:final_len],
    "transmission": trans_list[:final_len],
    "price": price_list[:final_len]
}

df = pd.DataFrame(data)
df.to_csv("cars24_data.csv", index=False)
print("Data saved to cars24_data.csv")
```
